**Creating a new column cost per unit and Profit Margin**

```python
dataset['Cost Per Unit'] = (dataset['Sales']- dataset['Profit'])/dataset['Quantity']
```

```python
dataset['Profit Margin'] = (dataset['Profit']/ dataset['Sales']) * 100
```

**Since the order date is in object so we convert it to date time**

```python
dataset['Date'] = pd.to_datetime(dataset['Order Date'], dayfirst= False)
dataset[' Order Month'] = dataset['Date'].dt.month_name()
dataset[' Order Day'] = dataset['Date'].dt.day_name()
dataset['Order Year'] = dataset['Date'].dt.year
```

**Since the dates have been separated now we drop the order date, ship date and Date and other unimportant columns**

```python
dataset.drop('Order Date', axis = 1, inplace = True)
dataset.drop('Ship Date', axis = 1, inplace = True)
dataset.drop('Date', axis = 1, inplace = True)
dataset.drop('Country', axis =1 , inplace = True)
dataset.drop('Postal Code', axis = 1, inplace= True)
```

Boxplot for sales, quantity and cost per unit to find the outliers

```python
plt.figure(figsize=(8, 5))
plt.boxplot(dataset['Sales'])
plt.title('Sales Distribution & Outliers')
plt.ylabel('Sales ($)')
plt.show()
```

 This shows there are many outliers for the sales column. The outliers can be seen due to  purchasing of high number of quantity for a product which directly affect the sales and shows outliers in them too.

```python
# Finding the outleirs

# For Sales

q1 = dataset['Sales'].quantile(0.25)
print(f'q1 is {q1}')
q3 = dataset['Sales'].quantile(0.75)
print(f'q3 is {q3}')
IQR = q3 - q1
print(f'The IQR is {IQR}')
lower_bound = q1 - 1.5 * IQR
print(f'The lower bound is {lower_bound}')
Upper_bound = q3 + 1.5 * IQR
print(f'The upper bound is {Upper_bound}')
Outliers = dataset[(dataset['Sales']< lower_bound)| (dataset['Sales']> Upper_bound)]
print(f'The outliers are:{Outliers}')
print(len(Outliers))

```

```python
# For Quantity
plt.figure(figsize=(8,5))
plt.boxplot(dataset['Quantity'])
plt.title('Box Plot For Quantity')
plt.ylabel('Quantity')
plt.show()

```

For Quantity there can be seen outliers it can be due to high number of qunatity brought by the consumers for a product. The outliers in quantity may be seen because of the bulk orders made by the corporate. Hence we do not remove the outliers.

```python
# Outliers

q1_q = dataset['Quantity'].quantile(0.25)
print(f'q1 is: {q1_q}')
q3_q = dataset['Quantity'].quantile(0.75)
print(f'q3 is: {q3_q}')
IQR_q = q3_q - q1_q
print(f'The IQR is: {IQR_q}')
lower_bound_q = q1_q - 1.5 * IQR_q
print(f'The lower bound is: {lower_bound_q}')
Upper_bound_q = q3_q + 1.5 * IQR_q
print(f'The upper bound is: {Upper_bound_q}')
Outliers_q = dataset[(dataset['Quantity']< lower_bound_q)| (dataset['Quantity']> Upper_bound_q)]
# print(f'The outliers are: {Outliers_q}')
print(len(Outliers_q))
```

There can be seen plenty of outliers that can be seen in both quantity and sales We did not remove them because these are real busienss transactions and not data entry errors.
Removing these ooutliers would result loss of the genuine data and accuracy for analysis which would affect the decisions.
